```python
# Sequential Search

def sqcsearch(data, key):
    pos = [];
    for i in range(len(data)):
        if data[i] == key:
            pos.append(i+1);
    if len(pos) > 0:
        print("data", key, "sebanyak", len(pos), "ditemukan di posisi", pos)
    else:
        print("data tidak ditemukan");
    return pos;

data = [8,3,3,5,7,4,3,1,7,12,5]
key = 3
result = sqcsearch(data,key)
```

```python
# Binary Search

def binsearch(data, key):
    awal = 1;
    akhir = len(data) + 1;
    ketemu = False;
    while (awal <= akhir) and not ketemu:
        tengah = int((awal + akhir) / 2);
        if key == data[tengah]:
            ketemu = True;
            print("data", key, "ditemukan di posisi", tengah);
        elif (key < data[tengah]):
            akhir = tengah - 1;
        else:
            awal = tengah + 1;
    if ketemu == False:
        print("data tidak ditemukan");

data = [2,3,3,4,5,6,7,7];
data.sort();
# Data nya di sorting dulu ascending

key = 3;
result = binsearch(data,key);
```

```python
def secsort(arr):
    l = len(arr);
    for i in range(l):    
        min_i = i
        
        # Ngecek apakah angka di index lebih besar dari angka lainnya     
        for j in range(i + 1, l):
            if arr[j] < arr[min_i]:
                min_i = j;
          
        # Tukeran Tempat     
        (arr[i], arr[min_i]) = (arr[min_i], arr[i]);
    return arr;

def binsearch(data, key):
    awal = 1;
    akhir = len(data) + 1;
    ketemu = False;

    while (awal <= akhir) and not ketemu:
        tengah = int((awal + akhir) / 2);
        if key == data[tengah]:
            ketemu = True;
            print("data", key, "ditemukan di posisi", tengah);
        elif (key < data[tengah]):
            akhir = tengah - 1;
        else:
            awal = tengah + 1;
    if ketemu == False:
        print("data tidak ditemukan");

data = [5,8,3,0,123,12,5,7,9,3,1,45,12,64,75];
data = secsort(data);

key = 75;
print(data);

result = binsearch(data,key);   
```
