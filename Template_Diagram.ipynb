```python
import matplotlib.pyplot as plt


#Data buah-buahan dan jumlah penjualan
buah = ['Apel', 'Jeruk', 'Pisang', 'Anggur']
penjualan = [50, 70, 90, 60]

# Membuat diagram batang
plt.bar(buah, penjualan, color='skyblue')

#Menambahkan judul dan label sumbu
plt.title('Diagram Batang Jumlah Penjualan Buah-Buahan')
plt.xlabel('Buah')
plt.ylabel('Jumlah Penjualan')

# Menampilkan diagram batang
plt.show()
```

```python
import matplotlib.pyplot as plt

#Data kategori
kategori = ['Menang', 'Seri', 'Kalah']
jumlah_pertandingan = [4, 3, 3]

#Hitung persentase
total_pertandingan = sum(jumlah_pertandingan)
persentase = [(jumlah / total_pertandingan) * 100 for jumlah in jumlah_pertandingan]

#Konstruksi diagram lingkaran
plt.figure(figsize=(6, 6))
plt.pie(persentase, labels=kategori, autopct='%1.1f%%', startangle=90)
plt.title('Diagram Lingkaran Jumlah Pertandingan')
plt.axis('equal') # Agar lingkaran menjadi Lingkaran sejati

plt.show()
```

```python
# HEATMAP
import matplotlib.pyplot as plt
import numpy as np

# Data korelasi antara variabel A dan B
data_korelasi = np.array([
    [0.8, 0.2, 0.5, 0.9],
    [0.6, 0.1, 0.3, 0.7],
    [0.3, 0.6, 0.4, 0.8]])

# Membuat heatmap
plt.imshow(data_korelasi, cmap='viridis', interpolation='nearest')

#Menambahkan label sumbu x dan y
plt.xticks(np.arange(len(data_korelasi[0])), ['A1', 'A2', 'A3', 'A4'])
plt.yticks(np.arange(len(data_korelasi)), ['B1', 'B2', 'B3'])

#Menambahkan nilai pada setiap sel
for i in range(len(data_korelasi)):
    for j in range(len(data_korelasi[0])):
        plt.text(j, i, '{:.2f}'.format(data_korelasi[i, j]),
                 ha='center', va='center', color='white')

#Menambahkan judul dan label sumbu
plt.title('Heatmap Korelasi antara Variabel A dan B')
plt.xlabel('Variabel A')
plt.ylabel('Variabel B')

#Menampilkan heatmap
plt.colorbar()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2

#Data himpunan A dan B
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

#Membuat Venn Diagram
venn2(subsets=(len(A - B), len(B - A), len(A & B)), set_labels=('A', 'B'))

#Menampilkan Venn Diagram
plt.title("Venn Diagram A dan B")
plt.show()
```

```python
import matplotlib.pyplot as plt

#Data keluhan pelanggan
keluhan = {
'Keterlambatan Pengiriman': 35,
'Produk Tidak Sesuai Pesanan': 20,
'Kualitas Produk Rendah': 15,
'Pelayanan Kurang Memuaskan': 10,
'Lainnya': 5
}

#Urutkan berdasarkan frekuensi dari yang tertinggi(param
keluhan_sorted = dict(sorted(keluhan.items(), key=lambda item: item[1], reverse=True))

#Hitung total frekuensi
total_frekuensi = sum(keluhan.values())

#Hitung persentase kumulatif
persentase_kumulatif = 0
persentase_kumulatif_list = []

for value in keluhan_sorted.values():
    persentase = (value / total_frekuensi) * 100
    persentase_kumulatif += persentase
    persentase_kumulatif_list.append(persentase_kumulatif)

# Gambar diagram Pareto
fig, ax1 = plt.subplots()

# Diagram batang
ax1.bar(keluhan_sorted.keys(), keluhan_sorted.values(), color='b')
ax1.set_xlabel('Jenis Keluhan')
ax1.set_ylabel('Frekuensi', color='b')

#Kurva kumulatif
ax2 = ax1.twinx()
ax2.plot(keluhan_sorted.keys(), persentase_kumulatif_list, color='r', marker='o')
ax2.set_ylabel('Persentase Kumulatif', color='r')

plt.title("Diagram Paretto")
plt.show()
```
